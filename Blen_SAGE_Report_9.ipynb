Finding of all partition of integer is too difficult and take a long time. But using sagemath can find easly. 

```python
par = int(input("Enter the Integer "))#To enter the integer 

#To find the partition of the integer
A = Partitions(par).cardinality()

#To print the partition of the integer
print("Total number of partitions for",par,"is",A) 

#To determine the even  the partition of the integer
C = list(Partitions(par, parts_in = [2,4..8]))

#To print the even partition
print("The even_parts of the partition are :- ",C)

#To determine the odd the partition of the integer
D = list(Partitions(par, parts_in = [1,3..9]))

#To print the odd partition
print("The even_parts of the partition are :- ",D)

#s = S.didtnict_element()
```
