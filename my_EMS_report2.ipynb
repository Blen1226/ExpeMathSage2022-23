Suppose we have coins of value $a=2$ cents and $b=3$ cents. What amount of money can we get by combining these coins?

we try to think about how can formulate the question  mathematically.
We use the formula $n = ax +by$ ,for all $x,y \geq 0$ and $1\leq n \leq 100$.

If $n=1$ can't get the money using the value of a and b.

```sage
n = 2
a = 2
b = 3
while n <= 100:
    for x in range (0,101):
        for y in range (0,101):
            if (a*x + b*y == n):
                print(x,y,n)
    n +=1            
```
